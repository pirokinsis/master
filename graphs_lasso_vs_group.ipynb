```python
import pandas as pd
```

```python
data=pd.read_csv("data.txt",delim_whitespace=True,names=['Lasso accuracy','Lasso accuracy pairs','Group lasso accuaracy pairs'])
```

```python
data
```

```python
data=data.sort_values(by=data.columns[1])
data.insert(0,"rank_pairs",range(0,data.shape[0]))
```

```python
data=data.sort_values(by=data.columns[1])
data.insert(0,"rank_lasso",range(0,data.shape[0]))
```

```python
data
```

```python
data.sort_values(by="rank_pairs").plot(x='rank_pairs',y=['Lasso accuracy','Lasso accuracy pairs','Group lasso accuaracy pairs'])
```

```python
data.sort_values(by="rank_lasso").plot(x='rank_lasso',y=['Lasso accuracy','Lasso accuracy pairs','Group lasso accuaracy pairs'])
```
